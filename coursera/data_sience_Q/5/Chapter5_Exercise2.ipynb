# Chapter 5 - Exercise 2: Merge dữ liệu

Hướng dẫn chung:
1. Tạo DataFrame từ dictionary:  <p style="color:blue">biến_dataframe = pd.DataFrame(<biến_dictionary>, columns = [danh sách tên cột])</p>
2. Gộp dữ liệu từ các DataFrame theo dòng: <span style="color:blue">biến_dataframe_mới = pd.concat([danh sách các dataframe cần gộp])</span>
3. Gộp dữ liệu từ các DataFrame theo cột: <span style="color:blue">biến_dataframe_mới = pd.concat([danh sách các dataframe cần gộp], axis = 1)</span>
4. Gộp dữ liệu từ 2 DataFrame theo cột chung: <span style="color:blue">biến_dataframe_mới = pd.merge(dataframe1, dataframe2, on = 'tên cột chung của 2 dataframe')</span>
