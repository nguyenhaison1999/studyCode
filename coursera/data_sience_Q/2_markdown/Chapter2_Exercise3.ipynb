# Chapter 2 : Sử dụng Markdown 


### Bài tập 2.3: Tạo một bảng chứa các môn học mà bạn yêu thích bao gồm các cột: stt, tên khóa học, thời lượng

#### cú pháp Markdown tạo Table sử dụng trong bài tập
<pre>
- ký hiệu | để phân cách các cột
- ký hiệu --- để phân cách dòng tiêu đề và dòng dữ liệu
- ký hiệu : để canh lề ( :--- canh trái; ---: canh phải; :---: canh giữa )
ví dụ:
|Tiêu đề cột 1|Tiều đề cột 2|Tiêu đề cột 3|
|:---|:---:|---:|
|nội dung 11|nội dung 12|nội dung 13|
|nội dung 21|nội dung 22|nội dung 23|
|nội dung 31|nội dung 32|nội dung 33|
</pre>
